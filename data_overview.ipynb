# Get an overview over the data
We use pymongo to get the data from the MongoDB and we use pandas to get a look at the data in a tabular format.

```python
import pymongo
import pandas as pd
```

```python
# Store all collections in a dictionary

client = pymongo.MongoClient('localhost')
channels = client.telegram.channels
entities = client.telegram.entities
errors = client.telegram.errors
messages = client.telegram.messages
users = client.telegram.users

collections = {"Channels" : channels, "Entities" : entities, "Errors" : errors, "Messages" : messages, "Users" : users}
```

```python
MAX_NUMBER_OF_SAMPLES = 30
dataframes = {}

# Transform the collections into dataframes with at most MAX_NUMBER_OF_SAMPLES rows
for label, collection in collections.items():
    cursor = collection.find().limit(MAX_NUMBER_OF_SAMPLES)
    entries = list(cursor)
    df = pd.DataFrame(entries)
    dataframes[label] = df
```
